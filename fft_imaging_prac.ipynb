```python
from package_nanonis.nanonis_sxm import * #nanonis_sxm 로드
import numpy as np 
import matplotlib.pyplot as plt
from package_utils.cmap_custom import nanox
from skimage.feature import peak_local_max
```

```python
a = topography('Topo/TNS_STM6_#117_2938.sxm')
z = a.get_z('subtract plane fit')*1e9
plt.imshow(z, vmin = np.mean(z) - 3*np.std(z), vmax = np.mean(z) + 3*np.std(z), cmap = nanox())
cbar = plt.colorbar(location='right') #colormap 추가
cbar.set_label('Z (nm)') # 컬러맵 이름 추가
plt.axis('off')
```

```python
two_d_fft_sqrt = fft.two_d_FFT_sqrt(z)
plt.imshow(two_d_fft_sqrt, vmin = np.mean(two_d_fft_sqrt) - 3*np.std(two_d_fft_sqrt), vmax = np.mean(two_d_fft_sqrt) + 3*np.std(two_d_fft_sqrt), cmap = nanox())
plt.axis('off')
```

```python
two_d_fft_log = fft.two_d_FFT_log(z)
plt.imshow(two_d_fft_log, vmin = np.mean(two_d_fft_log) - 3*np.std(two_d_fft_log), vmax = np.mean(two_d_fft_log) + 3*np.std(two_d_fft_log), cmap = nanox())
plt.axis('off')
```

```python
two_d_fft_lin = fft.two_d_FFT_lin(z)
plt.imshow(two_d_fft_lin, vmin = np.mean(two_d_fft_lin) - 3*np.std(two_d_fft_lin), vmax = np.mean(two_d_fft_lin) + 3*np.std(two_d_fft_lin), cmap = nanox())
plt.axis('off')
```

```python
# FFT image crop
Scanpixels = a.header['scan_pixels'] # 스캔 픽셀 수
center = round(Scanpixels[0]/2) # 원점 지정 (y픽셀/2)
croppx = 50 # crop 크기
two_d_fft_lin_crop = two_d_fft_lin[center-croppx:center+croppx, center-croppx:center+croppx] # 보여줄 픽셀 ( y축, x축 순)

lmax = peak_local_max(two_d_fft_lin_crop, min_distance=10)

# display results
fig, ax = plt.subplots(1, 2, figsize = (6,3), sharex=True, sharey=True)
ax[0].imshow(two_d_fft_lin_crop,  vmin = np.mean(two_d_fft_lin_crop) - 3*np.std(two_d_fft_lin_crop), vmax = np.mean(two_d_fft_lin_crop) + 3*np.std(two_d_fft_lin_crop), cmap = nanox())
ax[0].axis('off')
ax[0].set_title('Original')

ax[1].imshow(two_d_fft_lin_crop,  vmin = np.mean(two_d_fft_lin_crop) - 3*np.std(two_d_fft_lin_crop), vmax = np.mean(two_d_fft_lin_crop) + 3*np.std(two_d_fft_lin_crop), cmap = nanox())
ax[1].plot(lmax[:, 1], lmax[:, 0], 'r.')
ax[1].axis('off')
ax[1].set_title('Peak local max')

lmax
```
