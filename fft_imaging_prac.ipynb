```python
from package_nanonis.nanonis_sxm import * #nanonis_sxm 로드
import numpy as np 
import matplotlib.pyplot as plt
from package_utils.cmap_custom import nanox
```

```python
a = topography('Topo/TNS_STM6_#117_2938.sxm')
z = a.get_z('subtract plane fit')
plt.imshow(z, vmin = np.mean(z) - 3*np.std(z), vmax = np.mean(z) + 3*np.std(z), cmap = nanox()) 
plt.axis('off') 
```

```python
two_d_fft_sqrt = fft.two_d_FFT_sqrt(z)
plt.imshow(two_d_fft_sqrt, vmin = np.mean(two_d_fft_sqrt) - 3*np.std(two_d_fft_sqrt), vmax = np.mean(two_d_fft_sqrt) + 3*np.std(two_d_fft_sqrt), cmap = nanox())
plt.axis('off')
```

```python
two_d_fft_log = fft.two_d_FFT_log(z)
plt.imshow(two_d_fft_log, vmin = np.mean(two_d_fft_log) - 3*np.std(two_d_fft_log), vmax = np.mean(two_d_fft_log) + 3*np.std(two_d_fft_log), cmap = nanox())
plt.axis('off')
```

```python
two_d_fft_lin = fft.two_d_FFT_lin(z)
plt.imshow(two_d_fft_lin, vmin = np.mean(two_d_fft_lin) - 3*np.std(two_d_fft_lin), vmax = np.mean(two_d_fft_lin) + 3*np.std(two_d_fft_lin), cmap = nanox())
plt.axis('off')
```

```python
b = topography('Topo/Cu100_W_LN2_1686.sxm')
x = b.get_z('subtract linear fit xy')
plt.imshow(x, vmin = np.mean(x) - 3*np.std(x), vmax = np.mean(x) + 3*np.std(x), cmap = nanox()) 
plt.axis('off') 
```

```python
two_d_fft_sqrt = fft.two_d_FFT_sqrt(x)
plt.imshow(two_d_fft_sqrt, vmin = np.mean(two_d_fft_sqrt) - 3*np.std(two_d_fft_sqrt), vmax = np.mean(two_d_fft_sqrt) + 3*np.std(two_d_fft_sqrt), cmap = nanox())
plt.axis('off')
```

```python
two_d_fft_log = fft.two_d_FFT_log(x)
plt.imshow(two_d_fft_log, vmin = np.mean(two_d_fft_log) - 3*np.std(two_d_fft_log), vmax = np.mean(two_d_fft_log) + 3*np.std(two_d_fft_log), cmap = nanox())
plt.axis('off')
```

```python
# FFT image crop
Scanpixels = b.header['scan_pixels'] # 스캔 픽셀 수
center = round(Scanpixels[0]/2) # 원점 지정 (y픽셀/2)
croppx = 50 # crop 크기
two_d_fft_sqrt_crop = two_d_fft_sqrt[center-croppx:center+croppx, center-croppx:center+croppx] # 보여줄 픽셀 ( y축, x축 순)
plt.imshow(two_d_fft_sqrt_crop, vmin = np.mean(two_d_fft_sqrt_crop) - 3*np.std(two_d_fft_sqrt_crop), vmax = np.mean(two_d_fft_sqrt_crop) + 3*np.std(two_d_fft_sqrt_crop), cmap = nanox())
plt.axis('off')
```
