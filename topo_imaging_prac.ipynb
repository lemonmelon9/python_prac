```python
from package_nanonis.nanonis_sxm import * #nanonis_sxm 로드
import numpy as np 
import matplotlib.pyplot as plt
from package_utils.cmap_custom import nanox #custom color 로드
```

```python
a = topography('Topo/TNS_STM6_#117_2384.sxm') #topography 클래스
z = a.get_z() # 기본값 raw, fwd
```

```python
a.fname #파일 이름
```

```python
a.signals #시그널 데이터 전부 (딕셔너리 형태)
```

```python
a.signals ['Z']['forward'] # Z, forward에 대한 데이터 불러오기
```

```python
plt.imshow(z, aspect = 1024/256, vmin = np.mean(z) - 3*np.std(z), vmax = np.mean(z) + 3*np.std(z), cmap = nanox()) #aspect: 해상도
plt.axis('off') # x,y축 표시안함
# vmin and vmax define the data range that the colormap covers. +- 3시그마값 까지
```

```python
z_subav = a.get_z('subtract average')
plt.imshow(z_subav, aspect = 1024/256, vmin = np.mean(z_subav) - 3*np.std(z_subav), vmax = np.mean(z_subav) + 3*np.std(z_subav), cmap = nanox()) 
plt.axis('off') 
```

```python
z_sublf = a.get_z('subtract linear fit')
plt.imshow(z_sublf, aspect = 1024/256, vmin = np.mean(z_sublf) - 3*np.std(z_sublf), vmax = np.mean(z_sublf) + 3*np.std(z_sublf), cmap = nanox()) 
plt.axis('off') 
```

```python
z_sublfxy = a.get_z('subtract linear fit xy')
plt.imshow(z_sublfxy, aspect = 1024/256, vmin = np.mean(z_sublfxy) - 3*np.std(z_sublfxy), vmax = np.mean(z_sublfxy) + 3*np.std(z_sublfxy), cmap = nanox()) 
plt.axis('off') 
```
