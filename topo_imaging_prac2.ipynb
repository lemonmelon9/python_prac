```python
from package_nanonis.nanonis_sxm import * #nanonis_sxm 로드
import numpy as np 
import matplotlib.pyplot as plt
from package_utils.cmap_custom import nanox
```

```python
a = topography('Topo/TNS_STM6_#117_2938.sxm')
z_sublfxy = a.get_z('subtract linear fit xy')
plt.imshow(z_sublfxy, vmin = np.mean(z_sublfxy) - 3*np.std(z_sublfxy), vmax = np.mean(z_sublfxy) + 3*np.std(z_sublfxy), cmap = nanox()) 
plt.axis('off') 
```

```python
b = topography('Topo/TNS_STM6_#117_3782.sxm')
z = b.get_z('raw','bwd')
plt.imshow(z, vmin = np.mean(z) - 3*np.std(z), vmax = np.mean(z) + 3*np.std(z), cmap = nanox())
plt.axis('off')
```

```python
c = currentmap('Topo/TNS_STM6_#117_3782.sxm')
i_c = c.get_map()
plt.imshow(i_c, vmin = np.mean(i_c) - 4*np.std(i_c), vmax = np.mean(i_c) + 4*np.std(i_c), cmap = nanox())
plt.axis('off')
```
